```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
velocity = pd.read_csv('synthetic_data_generation/features.csv', header = None)
fric_coef = pd.read_csv('synthetic_data_generation/targets_AgingLaw.csv', header = None)
```

```python
# Flatten the velocity data to 1D for histogram
velocity_values = velocity.values.flatten()

# Define the bins for the log scale
logbins = np.logspace(np.log10(velocity_values.min()), np.log10(velocity_values.max()), 10)

# Plot the histogram with log scale
plt.figure(figsize=(10, 6))
plt.hist(velocity_values, bins=logbins)

# Set x-axis to logarithmic scale
plt.xscale('log')

# Title and labels
plt.title('Histogram of Velocity Values (Log Scale)')
plt.xlabel('Velocity')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Flatten the velocity data to 1D for histogram
velocity_values = np.log(np.log (1/np.array(velocity.values.flatten()) ) )
velocity_values = (velocity_values - np.min(velocity_values))/ (np.max(velocity_values) - np.min(velocity_values))

max_vel = velocity_values.max()
min_vel = velocity_values.min()

# Plot the histogram with log scale
plt.figure(figsize=(10, 6))
plt.hist(velocity_values, bins=10)

# Set x-axis to logarithmic scale
#plt.xscale('log')

# Title and labels
plt.title('Histogram of Velocity Values')
plt.xlabel('Velocity')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plot
plt.show()
```
